a) Encontrar un conjunto de datos y elegir al menos 4 variables continuas

Se selecciona un dataset que tiene variables continuas en las columnas por edad cuales son sus activos, ganancias, salarios y ventas 

```python
# Dataset
import pandas as pd
dataset = "/Users/cristianlopez/Downloads/Ejemplo_SalarioMultiple.csv"
dataset = pd.read_csv(dataset)
dataset
```

```python
dataset.dtypes
```

```python
#Seleccionar variables continuas
variables_continuas = dataset [["assets","profits","roa","sales"]]
variables_continuas
```

b) Realizar gráfica de dispersión por pares de variables. Calcular y representar el vector de medias

```python
import matplotlib.pyplot as plt

# Calcular medias de profits y otras variables
mean_profits = variables_continuas["profits"].mean()
mean_assets = variables_continuas["assets"].mean()
mean_roa = variables_continuas["roa"].mean()
mean_sales = variables_continuas["sales"].mean()

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(variables_continuas["profits"], variables_continuas["assets"], label="Assets")
plt.scatter(variables_continuas["profits"], variables_continuas["roa"], label="ROA")
plt.scatter(variables_continuas["profits"], variables_continuas["sales"], label="Sales")

# Dibujar vectores de medias
plt.arrow(mean_profits, mean_assets, 0, 0, color='r', width=0.05, head_width=0.3, alpha=0.5, label="Mean Vectors")
plt.arrow(mean_profits, mean_roa, 0, 0, color='r', width=0.05, head_width=0.3, alpha=0.5)
plt.arrow(mean_profits, mean_sales, 0, 0, color='r', width=0.05, head_width=0.3, alpha=0.5)

# Etiquetas y título del gráfico
plt.xlabel("Profits")
plt.ylabel("Values")
plt.title("Gráfico de Dispersión de Profits vs Variables Continuas con Vectores de Medias")
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico

```
